```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
doc_types = [i for i in range(100)]
```

```python
def prepare_data(load_am):
    data = []
    for i in range(load_am):
        dtype = np.random.randint(0, len(doc_types))
        data.append((doc_types[dtype], np.random.rand()*dtype+dtype, np.random.rand()*dtype+dtype))
    return data
```

```python
data = prepare_data(10000)
```

```python
df = pd.DataFrame(data, columns=['doc_type','x','y'])
```

```python
def pick_colors(number):
    x = np.linspace(0, 1, 10)
    cmap = plt.get_cmap('gnuplot')
    colors = [cmap(i) for i in np.linspace(0, 1, number)]
    return colors
```

```python
colors = pick_colors(len(doc_types))
```

```python
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5, forward=True)

color_labels = df['doc_type'].unique()

# List of RGB triplets
rgb_values = sns.color_palette("Set2", 8)

# Map label to RGB
color_map = dict(zip(color_labels, colors))

# Finally use the mapped values
plt.scatter(df['x'], df['y'], c=df['doc_type'].map(color_map), s=10, alpha=0.5)
```
